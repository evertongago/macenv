```python
import six
import sys
sys.modules['sklearn.externals.six'] = six

import pandas, mlrose, numpy

SEED = 42
MONEY = 5.0
numpy.random.seed(SEED)
```

```python
df = pandas.read_csv('data/prices.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(['LVTC3'], axis=1, inplace=True)
print(f'Columns {len(df.columns)}')
```

```python
# https://www.bcb.gov.br/controleinflacao/historicotaxasjuros

selic = numpy.array([
    11.75, # 17/03/2022 -
    10.75, # 03/02/2022 - 16/03/2022
    9.25,  # 09/12/2021 - 02/02/2022
    7.75,  # 28/10/2021 - 08/12/2021
    6.25,  # 23/09/2021 - 27/10/2021
    5.25,  # 05/08/2021 - 22/09/2021
    4.25,  # 17/06/2021 - 04/08/2021
    3.50,  # 06/05/2021 - 16/06/2021
    2.75,  # 18/03/2021 - 05/05/2021
    2.00   # 21/01/2021 - 17/03/2021
    
])
```

```python
def fit(solve):
    _df = df.copy()
    weights = solve / solve.sum()
    
    # Normalize values
    for i in df.columns[1:]:
        _df[i] = _df[i] / _df[i][0]
    
    # Allocate money on stocks
    for i, stock in enumerate(_df.columns[1:]):
        _df[stock] = _df[stock] * weights[i] * MONEY

    # Drop column
    _df.drop(labels=['Date'], axis=1, inplace=True)
    
    _df['accum'] = _df.sum(axis=1)
    
    # Calculate return by stock
    _df['rate'] = ((_df['accum'] / _df['accum'].shift(1)) - 1) * 100
    
    # Calculate sharpe ration
    sharpe_ratio = (_df['rate'].mean() - (selic.mean() / 100)) / (_df['rate'].std() * numpy.sqrt(246))
    
    return sharpe_ratio

fn = mlrose.CustomFitness(fit)
optimizer = mlrose.ContinuousOpt(length=len(df.columns[1:]), fitness_fn=fn, maximize=True, min_val=0, max_val=1)
```

```python
solve, _ = mlrose.hill_climb(optimizer, random_state=SEED)
solve = solve / solve.sum()

solve
```

```python
columns = []
amount = []

def view(solve):
    cols = df.columns[1:]
    indexes = numpy.argsort(solve)[::-1]
    for i in indexes:
        if solve[i] <= 0:
            break

        if (solve[i] > 0.01):
            columns.append(cols[i])
            amount.append(solve[i])
    
    for c, a in zip(columns, amount):
        print('Stock: {} - {:.1f}% = R$ {:.2f}'.format(c, a * 100, MONEY * a * 1000))

        
view(solve)
```

```python
def backtesting(df, money, weights):
    _df = df.copy()
    
    # Normalize values
    for i in df.columns[1:]:
        _df[i] = _df[i] / _df[i][0]
    
    # Allocate money on stocks
    for i, stock in enumerate(_df.columns[1:]):
        _df[stock] = _df[stock] * weights[i] * money

    # Drop column
    _df.drop(labels=['Date'], axis=1, inplace=True)
    
    _df['accum'] = _df.sum(axis=1)
    
    return _df.loc[len(_df) - 1]['accum']

# Run backtest
print('Start with: R$ {:.2f} Current: R$ {:.2f}'.format(MONEY, backtesting(df=df, money=MONEY, weights=solve)))
```

```python
df_norm = df[['Date'] + columns].copy()
for i in df_norm.columns[1:]:
    df_norm[i] = df_norm[i] / df_norm[i][0]

df_norm.plot(x='Date', figsize=(15, 7), title='Prices History')
```
