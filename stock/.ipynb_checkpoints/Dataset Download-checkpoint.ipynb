```python
import pandas
from pandas_datareader import data
```

```python
df = pandas.read_csv('data/fundamental_trusted.csv')
df.head()
```

```python
df_filtered = df[df['LIQ_2MESES'] > 1000000]

df_ranking = pandas.DataFrame()
df_ranking['IDX'] = range(0, 50)
df_ranking['EV/EBIT'] = df[df['EV/EBIT'] > 0].sort_values(by=['EV/EBIT'])['PAPEL'][:50].values
df_ranking['ROIC'] = df.sort_values(by=['ROIC'], ascending=False)['PAPEL'][:50].values

df_ranking.head()
```

```python
df_a = df_ranking.pivot_table(columns='EV/EBIT', values='IDX')
df_b = df_ranking.pivot_table(columns='ROIC', values='IDX')

# Merge dataframes
df_m = pandas.concat([df_a, df_b])

df_rank = df_m.dropna(axis=1).sum()
df_rank.sort_values()
```

```python
stocks = df_rank.index.tolist()
df = pandas.DataFrame()
for i in stocks:
    df[i] = data.DataReader(name=f'{i}.SA', data_source='yahoo', start='2021-03-15')['Close']
    
df.to_csv('data/prices.csv', sep=',', header=True, index=True)
```
