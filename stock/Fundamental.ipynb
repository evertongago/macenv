```python
import warnings, pandas

warnings.filterwarnings('ignore')
```

```python
df_fund = pandas.read_csv('data/fundamental.csv', decimal=',', thousands='.')

for c in ['DIV_YIELD', 'MRG_EBIT', 'MRG_LIQ', 'ROIC', 'ROE', 'CRESC_REC_5A']:
    df_fund[c] = df_fund[c].str.replace('.','')
    df_fund[c] = df_fund[c].str.replace(',','.')
    df_fund[c] = df_fund[c].str.rstrip('%').astype(float) / 100

df_fund.head()
```

#### Agro

```python
df_agro = pandas.read_csv('data/agro.csv')
df_agro['SETOR'] = 'Agro'

df_fund = pandas.merge(df_fund, df_agro, left_on='PAPEL', right_on='PAPEL', how='left')

df_fund.head()
```

#### Comercio & Distribuição

```python
df_com = pandas.read_csv('data/comdist.csv')
df_com['SETOR'] = 'Comércio & Distribuição'

df_fund = pandas.merge(df_fund, df_com, left_on='PAPEL', right_on='PAPEL', how='left')

df_fund['SETOR_x'] = df_fund.apply(lambda d : d['SETOR_y'] if pandas.isna(d['SETOR_x']) else d['SETOR_x'], axis=1)
df_fund.drop(labels=['SETOR_y'], axis=1, inplace=True)
df_fund.rename(columns={'SETOR_x': 'SETOR'}, inplace=True)


df_fund.head()
```
