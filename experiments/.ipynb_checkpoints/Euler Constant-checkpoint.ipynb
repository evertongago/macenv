```python
def fatorial(n: int) -> int:
    if n == 1:
        return 1
    
    return n * fatorial(n - 1)
```

```python
fatorial(5)
```

```python
e = 1
v = []
for i in range(1, 10):
    e = e + (1 / fatorial(i))
    v.append(e)
    print('{:.5f}'.format(e))
```

```python
import math

k = []
m = []
for i in range(1, 30 + 1):
    e = math.pow(1 + (1 / i), i)
    m.append(e)
    if i % 10 == 0:
        k.append('{:.5f}'.format(e))
    
print(k)
```

```python
from matplotlib import pyplot
pyplot.plot(range(1, 30 + 1), m)
```
